```python
response = requests.get(url, verify=False)
```

```python
import requests
from bs4 import BeautifulSoup


# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    # Parsear el contenido HTML
    soup = BeautifulSoup(response.text, 'html.parser')

    # Ahora puedes usar el objeto soup para encontrar elementos en el HTML
    summary_div = soup.find('div', {'id': 'shiny-tab-summary'})

    # Resto de tu código aquí...
else:
    print('Error al obtener la página:', response.status_code)
```

```python
if response.status_code == 200:
    html = response.text
    summary_div = soup.find('div', {'id': 'shiny-tab-summary'})

    data = {
    'Texto': [],
    'Numero': []}

    if summary_div:
        relevant_divs = summary_div.find_all('div', {'class': 'col-sm-4'})
        div = relevant_divs[0]
        cubos = []
        for div in relevant_divs:
          p_elements = div.find_all('p')

          if (len(p_elements)) > 3:
            cubos.append(p_elements[:2])
            cubos.append(p_elements[3:5])

          else:
            p_element = p_elements[:2]
            cubos.append(p_element)

        for cubo in cubos:
          texto = cubo[1].get_text(strip=True)
          data['Texto'].append(texto)
          numero = cubo[0].get_text(strip=True)
          data['Numero'].append(numero)

    tablero = pd.DataFrame(data)

else:
    print('Error al obtener la página:', response.status_code)
```

```python

# Abre el archivo en modo binario de lectura ('rb')
with open('/content/pdf2.pdf', 'rb') as archivo:
    # Crea un objeto PdfFileReader
    lector = PyPDF2.PdfReader(archivo)

    # Inicializa una cadena vacía para almacenar el texto
    texto = ''

    # Itera sobre todas las páginas del PDF
    for i in range(len(lector.pages)):
        # Obtiene la página
        pagina = lector.pages[i]

        # Extrae el texto de la página y lo añade a la cadena de texto
        texto += pagina.extract_text ()

# Imprime el texto extraído
print(texto)
```
